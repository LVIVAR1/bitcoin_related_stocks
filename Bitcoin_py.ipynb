```python
bitcoin = pd.read_csv('/Users/luisvivar/Desktop/BTC-USD.csv')
mara = pd.read_csv('/Users/luisvivar/Desktop/MARA.csv')
riot = pd.read_csv('/Users/luisvivar/Desktop/RIOT.csv')
coinbase = pd.read_csv('/Users/luisvivar/Desktop/COIN.csv')
hood = pd.read_csv('/Users/luisvivar/Desktop/HOOD.csv')
```

```python
dfs = [bitcoin, mara, riot, coinbase, hood]
```

```python
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
bitcoin.head(200)
```

```python
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)
```

```python
bitcoin.head()
```

```python
for df in dfs:
    df['Change in price'] = df['Close'] - df['Previous day close price']
```

```python
bitcoin.head()
```

```python
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```

```python
bitcoin.head()
```

```python
for df in dfs:
    df['Previous day volume'] = df.Volume.shift(1)
```

```python
bitcoin.head()
```

```python
for df in dfs:
    df['Change in Volume'] = df['Volume'] - df['Previous day volume']
```

```python
bitcoin.head()
```

```python
for df in dfs:
    df['Percent change in volume'] = df.Volume.pct_change()
```
